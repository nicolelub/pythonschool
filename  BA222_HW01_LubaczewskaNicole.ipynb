```python
#Problem 2: Product Lookup

def ProductSearch(word, product, quantity):
        '''a product search'''
        if (word in product):
            n = len(product)
            x = 0
            for i in range(0,n):
                if ((word == product[i]) & (quantity[i] != 0)): 
                    x = i
                    print('The amount of', word, 'is', quantity[i])
                elif ((word == product[i]) & (quantity[i] == 0)):
                    print ('Error!')
        else:
            print('Error!')

product = ['tomatoes','potatoes','carrots','cucumber','zucchini']

quantity = [3,5,4,1,0]

ProductSearch('zucchini',product, quantity)
```

```python
#Problem 3

def OrderCalculation(items, prices, name, order, quantities):
    subtotal = 0 
    for i in order:
        item_position = items.index(i)
        p = prices[item_position]
        
        n = name[item_position]
        
        item_position = order.index(i)
        q = quantities[item_position]
        
        subtotal = subtotal + p*q 
        print(n, '(x', q, ')', p*q)
    
    print('Total', subtotal)
    
items = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
prices = [5.00,7.00,6.00,7.50,2.00,2.00,3.00,3.50]
name = ['Regular Burger','Regular Burger with Bacon','Cheeseburger', 'Bacon Cheeseburger','Coke','Sprite','Fries','Seasoned Fries']

order = ['A', 'E', 'H']
quantities = [2,1,1]

        
OrderCalculation (items, prices, name, order, quantities)
```

```python
# Problem 4 Continuation
#revenue based on avarage q demanded for each P


# Variable Cost function

# Fixed Cost 

fc = 50000
p = 20

def demand(p):
    return( qp[qp.price == p]['quantity demanded'].mean() )

def VarCost (q):
    q = demand(p)
    if (q < 15000):
        return(14 * q)
    elif ((q >= 15000) & (q < 20000)):
        return(12 * q)
    elif (q >= 20000):
        return(5 * q)

# revenue = [(5 * 20444.68,10 * 18886.07), (15 * 15021.190476) ,(20 * 9839.683168) , (25 * 5040.962069)]   
#VarCost = [(VarCost (20444.684211)),(VarCost (18886.068966)),(VarCost (15021.190476)), (VarCost (9839.683168)), (VarCost (5040.962069))]

def tc(p):
    return (VarCost(p)+fc)

def revenue(p):
    q = demand(p)
    return (p*q)

def profit(p):
    return(revenue(p) - tc(p))

profit(15)

#Highest profit for 15 after cleaning data!
```
