```python
import pandas as pd
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
path = '/Users/nicolelubaczewska/Downloads/ba222Project_CleanData_12December.csv'

cs = pd.read_csv(path)

cs.head()
```

```python
cs.Food_spending.astype('float')

smf.ols('Food_spending ~ C(Mealbox)', data = cs).fit().params
```

```python
reg1 = smf.ols('Food_spending ~ C(Mealbox)', data = cs).fit()

reg2 = smf.ols('Food_spending ~ C(Mealbox) + C(Allowance_month)', data = cs).fit()

reg3 = smf.ols('Food_spending ~ C(Mealbox) + Breakfast', data = cs).fit()

reg4 = smf.ols('Food_spending ~ C(Mealbox) + Dinner', data = cs).fit()

reg5 = smf.ols('Food_spending ~ C(Mealbox) + Starbucks', data = cs).fit()

reg6 = smf.ols('Food_spending ~ C(Mealbox) + CityCo', data = cs).fit()

reg7 = smf.ols('Food_spending ~ C(Mealbox) + Meals_Amount', data = cs).fit()

reg8 = smf.ols('Food_spending ~ C(Mealbox) + Age', data = cs).fit()

reg9 = smf.ols('Food_spending ~ C(Mealbox) + Living', data = cs).fit()


from statsmodels.iolib.summary2 import summary_col



summary_col([reg1, reg2, reg3, reg4, reg5, reg6, reg7, reg8, reg9],
            stars = True,
            regressor_order = ['Intercept', 'time'])


```
